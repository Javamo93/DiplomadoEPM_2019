## Application: Colombian electricity spot price

```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
%matplotlib inline

data_set = pd.read_csv('local/data/price.csv', sep = '\t', squeeze = True)
Prices = data_set.values
date_begin = np.datetime64('2000-01-01T00:00:00')
date_end = np.datetime64('2019-01-01T00:00:00')
Time = np.arange(date_begin,date_end,dtype='datetime64[h]')

plt.plot(Time,Prices)
plt.title("Hourly spot electricity price")
plt.ylabel('COL/kWh')
plt.show()
```

```python
from pandas import Series
from statsmodels.graphics.tsaplots import plot_acf
k=72
series = Series(Prices)
plot_acf(series, lags=k)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(series, lags=k)
plt.show()
```

```python
df = pd.DataFrame({'prices':Prices,'dates':Time})
#df['dates'] = pd.DatetimeIndex(df['dates'],freq ='H')
df['dates'] = pd.to_datetime(df['dates'])
df['just_date'] = df['dates'].dt.date
df['day_of_week'] = df['dates'].dt.weekday_name
df['month'] = df['dates'].dt.strftime('%b')
df['Year'] = df['dates'].dt.year
print(df[:100:25])
```

Obtengo los índices del último mes que se usará para realizar las pruebas de validación

```python
LastMont = df['month'][-1:].values
LastYear = df['Year'][-1:].values
df2 = df.loc[(df['month'] == LastMont[0]) & (df['Year'] == LastYear[0])]
IndLastMonth = df2.index[0]
print(df2[:100:25])
```

Como ya se explicó en sesiones anteriores filtramos por día:

```python
import local.lib.EnergyPricesLibrary as Ep
look_back = 1
Days = np.array(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
fig, axs = plt.subplots(4, 2, sharex='all')
fig.set_size_inches(20,20)
k1 = 0
k2 = 0
for j,d in enumerate(Days,0):
    trainX, trainY, testX, testY, scaler, df2, dataset = Ep.SplitTimeseries(df,'prices',day = d, ValData = 'index', TimeAhead = IndLastMonth, look_back = look_back)
    testY = scaler.inverse_transform(testY.reshape(len(testY),1))
    Nt = len(testY)
    k1 = j
    if j > 3:
        k1 = j-4
        k2 = 1
    for i in range(int(Nt/24)):
        Time3 = df2['just_date'][i*24:(i+1)*24].values
        axs[k1,k2].plot(testY[i*24:(i+1)*24,0],label=Time3[1])
        axs[k1,k2].set_title(d)
        axs[k1,k2].legend()
        axs[k1,k2].set_ylabel('COL/kWh')
```

```python
day = 'Monday'
df2 = df.loc[df['day_of_week'] == day]
plt.plot(df2['dates'],df2['prices'])
plt.title(day)
plt.ylabel('COL/kWh')
plt.show()
```

Las mismas transformaciones vistas en las sesiones anteriores pueden ser aplicadas en este caso, aunque la red neuronal es un modelo no lineal y no algunas de ellas pueden no ser necesarias.

## Box-Cox

```python
from scipy import stats
fig = plt.figure()
ax1 = fig.add_subplot(211)
x = stats.loggamma.rvs(5, size=500) + 5
prob = stats.probplot(df2['prices'], dist=stats.norm, plot=ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
ax2 = fig.add_subplot(212)
xt, lmbda = stats.boxcox(df2['prices'].values)
df2 = df2.assign(BoxCox = xt)
prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()
```

# Predicción de un paso adelante

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU
from sklearn.metrics import mean_squared_error
import math
look_back = 1
# TimeAhead = 96
trainX, trainY, testX, testY, scaler, dfval, dataset = Ep.SplitTimeseries(df,'prices',day = day, ValData = 'index', TimeAhead = IndLastMonth, look_back = look_back)
# create and fit the GRU network
model = Sequential()
model.add(GRU(10, input_shape=(1, look_back)))
model.add(Dense(1))
mean_squared_error
#model.compile(loss='mean_squared_error', optimizer='adam')
model.compile(loss='mean_squared_logarithmic_error', optimizer='adam')
#model.compile(loss='mean_absolute_error', optimizer='adam')
model.fit(trainX, trainY, epochs=10, batch_size=20, verbose=2)
```

```python
# make predictions
Predict = model.predict(np.concatenate((trainX,testX),axis=0))
```

```python
# invert predictions
trainPredict = scaler.inverse_transform(Predict[:trainX.shape[0],:])
trainY = scaler.inverse_transform(trainY.reshape(len(trainY),1))
testPredict = scaler.inverse_transform(Predict[trainX.shape[0]:,:])
testY = scaler.inverse_transform(testY.reshape(len(testY),1))
# calculate root mean squared error
trainScore = math.sqrt(mean_squared_error(trainY, trainPredict))
print('Train Score: %.2f RMSE' % (trainScore))
print('Train Score: %.2f MAPE' % Ep.MAPE(trainPredict,trainY))
testScore = math.sqrt(mean_squared_error(testY, testPredict))
print('Test Score: %.2f RMSE' % (testScore))
print('Test Score: %.2f MAPE' % Ep.MAPE(testPredict,testY))
```

```python
datasetO = scaler.inverse_transform(dataset)
Nt = trainPredict.shape[0] + testPredict.shape[0]+look_back
trainPredictPlot = np.zeros((Nt,1))
trainPredictPlot[:,:] = np.nan
trainPredictPlot[:len(trainPredict)+look_back, :] = np.concatenate((datasetO[:look_back].reshape(look_back,1),trainPredict))
# shift test predictions for plotting
testPredictPlot = np.zeros((Nt,1))
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+look_back:Nt, :] = testPredict
# plot baseline and predictions
plt.plot(scaler.inverse_transform(dataset),label='Original data')
plt.plot(trainPredictPlot,label='Training predictions')
plt.plot(testPredictPlot,label='Test prediction')
plt.legend()
plt.show()
```

```python
df3 = dfval.loc[(dfval['day_of_week'] == day) & (dfval.index >= IndLastMonth)]
Nt = len(df3)
plt.figure(figsize=(8,6))
for i in range(int(Nt/24)):
    Time3 = df3['just_date'][i*24:(i+1)*24].values
    plt.plot(testPredict[i*24:(i+1)*24,0],label=('Pred-'+ str(Time3[1])))
    plt.plot(testY[i*24:(i+1)*24,0],linestyle='dashed',label=Time3[1])
    plt.title(day)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.ylabel('COL/kWh')
plt.show()
```

### Removiendo la estacionalidad

```python
look_back = 1
trainX, trainY, testX, testY, scaler, dfval, dataset = Ep.SplitTimeseries(df2,'DesSeasonality',day = day, ValData = 'index', TimeAhead = IndLastMonth, look_back = look_back)
# create and fit the LSTM network
model = Sequential()
model.add(GRU(10, input_shape=(1, look_back)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=10, batch_size=50, verbose=2)
```

```python
# make predictions
Predict = scaler.inverse_transform(model.predict(np.concatenate((trainX,testX),axis=0)))
Predict = Predict + df2['BoxCox'][look_back:-24].values.reshape(-1,1)
trainY = scaler.inverse_transform(trainY.reshape(-1,1)) + df2['BoxCox'][look_back:trainX.shape[0]+look_back].values.reshape(-1,1)
testY = scaler.inverse_transform(testY.reshape(-1,1)) + df2['BoxCox'][-24-testX.shape[0]:-24].values.reshape(-1,1)
```

```python
# invert predictions
trainPredict = inv_boxcox(Predict[:trainX.shape[0],:], lmbda)
trainY = inv_boxcox(trainY.reshape(len(trainY),1), lmbda)
testPredict = inv_boxcox(Predict[trainX.shape[0]:,:], lmbda)
testY = inv_boxcox(testY.reshape(len(testY),1), lmbda)
# calculate root mean squared error
trainScore = math.sqrt(mean_squared_error(trainY, trainPredict))
print('Train Score: %.2f RMSE' % (trainScore))
print('Train Score: %.2f MAPE' % Ep.MAPE(trainPredict,trainY))
testScore = math.sqrt(mean_squared_error(testY, testPredict))
print('Test Score: %.2f RMSE' % (testScore))
print('Test Score: %.2f MAPE' % Ep.MAPE(testPredict,testY))
```

```python
df3 = dfval.loc[(dfval['day_of_week'] == day) & (dfval.index >= IndLastMonth)]
Nt = len(df3)
plt.figure(figsize=(8,6))
for i in range(int(Nt/24)):
    Time3 = df3['just_date'][i*24:(i+1)*24].values
    plt.plot(testPredict[i*24:(i+1)*24,0],label=('Pred-'+ str(Time3[1])))
    plt.plot(testY[i*24:(i+1)*24,0],linestyle='dashed',label=Time3[1])
    plt.title(day)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.ylabel('COL/kWh')
plt.show()
```
