**Example**: Diabetes data (https://www.kaggle.com/uciml/pima-indians-diabetes-database)

6. **mass**: Body mass index -BMI- (weight in kg/(height in m)^2)

9. **clas**: Outcome class variable (0 or 1) 268 of 768 are 1, the others are 0

# Concatenation
Refer for details to https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

```python
import pandas as pd

data = data.loc[5:10, :]
print(data.shape)
print(data)

D1 = pd.concat([data, data], axis=0, ignore_index=False) # concatenate DataFrames along rows
print(D1.shape)
print(D1)

D2 = pd.concat([data, data], axis=0, ignore_index=True)
print(D2.shape)
print(D2)

D3 = pd.concat([data, data], axis=1)                     # concatenate DataFrames along columns
print(D3.shape)
print(D3)
id = [i for i in range(0, D3.shape[0])] # or id = [i for i in range(1, D3.shape[0]+1)]
D3.index = id
D3
```

# More about concatenation

```python
import pandas as pd
print(data.shape)
print(data)
data1 = data.sample(frac=0.3)
print(data1.shape)
print(data1)

data2 = pd.concat([data, data1], axis=0)                      # Concatenate rows (same as rbind() in R)
print(data2.shape)
print(data2)
data3 = pd.concat([data, data1], axis=0, ignore_index=True)   # Concatenate rows (same as rbind() in R)
print(data3.shape)
print(data3)

data4 = pd.concat([data, data1], axis=1)                      # Concatenate columns (similar to cbind() in R)
print(data4.shape)
print(data4)

data5 = pd.concat([data.reindex(data1.index), data1], axis=1) # Concatenate columns (similar to cbind() in R)
print(data5.shape)
print(data5)

data6 = pd.concat([data, data1], axis=1, join='inner')        # Concatenate columns (similar to cbind() in R)
print(data6.shape)
data6
```

# Time series (time dependent data)

```python
import pandas as pd
import numpy as np

data = pd.read_csv('price.csv', sep = '\t', squeeze = True)
print(data)

p = data.values
index = pd.date_range('2000-01-01', periods = len(data), freq = 'H')

price = pd.DataFrame({'price':p}, index = index)
print(price)

# also
start = np.datetime64('2000-01-01T00:00:00')
end   = np.datetime64('2019-01-01T00:00:00')
t = np.arange(start, end, dtype = 'datetime64[h]')

print(t.shape)

price = pd.DataFrame({'price':p}, index = t)
#price = pd.DataFrame(data)
#price.reindex(index)
print(price)

import matplotlib.pyplot as mpl

mpl.plot(price, label='price')
mpl.plot(t,p, label='p')
mpl.title('Hourly spot electricity price')
mpl.ylabel('COL/kWh')
mpl.xlabel('Time')
mpl.legend(loc='upper left',)
mpl.show()
```

# Sequences

```python
# Sequences in Python are zero-indexed
seq = [i for i in range(20)]
print(seq)

# also
seq = range(0, 20, 1)
print(list(seq))

# Sequence from 1 to 20
seq = [i+1 for i in range(20)]
print(seq)
#print(type(seq))

# also
seq = range(1, 21, 1)
#print(seq)
print(list(seq))

d = pd.Series(seq) # seq as a Series (1D DataFrame)
print(d)
```

#  Extract a subset of a sequence

```python
# seq[start:end:increment] 
seq[1]        # evaluate at [1]
seq[2:10]     # evaluate from 2 to 10 (endpoints aren't included)
seq[2:10:2]   # evaluate from 2 to 8 by 2
seq[:4]       # evaluate at the first 4 elements of the list
seq[-2:]      # evaluate at the last two elements
seq[:]        # evaluate at all elements of the list (an exact copy of seq)
seq[::3]      # evaluate every third element
```
