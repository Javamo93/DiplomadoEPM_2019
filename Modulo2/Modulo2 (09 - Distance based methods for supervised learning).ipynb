## KNeighbours classification

this is **not parametric**, why?

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,y)
mlutils.plot_2Ddata_with_boundary(knn.predict, X, y)
```

observe what happens when we use different number of neighbours

```python
def plot_KNN(knn):
    plt.figure(figsize=(20,3))
    for i,n in enumerate([2,5,10,20,50]):
        print (".", end=" ")
        plt.subplot(1,5,i+1)
        knn.set_params(n_neighbors=n)
        knn.fit(X,y)
        mlutils.plot_2Ddata_with_boundary(knn.predict, X, y)
        plt.title("n neighbours = %d"%n)
```

```python
knn = KNeighborsClassifier()
plot_KNN(knn)
```

## Observe now what happens if we use our own distance metrics

```python
X,y = make_moons(100, noise=0.3)
```

```python
mlutils.plot_2Ddata(X,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def my_metric1(x,y):
    return np.abs(x[0]-y[0]) + 2.5*np.abs(x[1]-y[1])

knn = KNeighborsClassifier(metric=my_metric1, n_neighbors=20)
knn.fit(X,y)
mlutils.plot_2Ddata_with_boundary(knn.predict, X, y)
```

```python
plot_KNN(knn)
```

```python
def my_metric0(x,y):
    return np.abs(x[0]-y[0]) + .1*np.abs(x[1]-y[1])
knn = KNeighborsClassifier(metric=my_metric0, n_neighbors=20)
knn.fit(X,y)
mlutils.plot_2Ddata_with_boundary(knn.predict, X, y)
```

```python
plot_KNN(knn)
```
