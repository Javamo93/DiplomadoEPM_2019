**Qué es clustering?**

- **Objetivo**: agrupar objetos físicos o obstractos en clases de objetos **similares**
- Es una tarea **NO SUPERVISADA** $\rightarrow$ no sabemos a priorí cómo clasificar nuestros objetos
- Es una tarea **NO COMPLETAMENTE DEFINIDA** $\rightarrow$ ¿Cómo cuantificamos el desempeño de un resultado de clustering?


- ¿Qué definición de **similitud** establecemos?

**Ejemplos de aplicaciones de clustering**

- Taxonomías en biología, agrupaciones por similitud biológica, o incluso genética (big data!!)
- Páginas similares para estructurar resultados de búsquedas (p.ej. La búsqueda de "película" podría devolver resultados agrupados por descripciones similares.
- Segmentación de clientes o usuarios por un criterio de similitud definido.

pseudo código

    input: 
        X: datos
        k: número de clusters deseados
        
    algoritmo:
        1. selecciona k centroides aleatoriamente
        2. repite hasta que los k centroides no cambien:
        3.     establece k clusters asignado cada dato al centroide más cercano
        4.     recalcula el centroide de cada cluster como el promedio de los datos

## Métodos basados en centroides -  KMeans


```python
class KM:
    def __init__(self, n_clusters, metric=None):
        self.n_clusters = n_clusters
        if metric == None:
            self.metric = lambda x,y: np.linalg.norm(x-y)
        else:
            self.metric = metric
            
    def fit(self, X):
        cluster_assignment = None
        last_cluster_assignment = None
        clusters = np.random.random(size=(self.n_clusters, X.shape[1]))
        clusters = clusters * (X.max(axis=0)-X.min(axis=0)) + X.min(axis=0)

        while cluster_assignment is None or last_cluster_assignment is None \
              or not np.allclose(cluster_assignment, last_cluster_assignment):

            last_cluster_assignment = cluster_assignment
            
            # assign each data point to closest centroid
            d = np.r_[[[self.metric(c, x) for x in X] for c in clusters]].T
            cluster_assignment = np.argmin(d, axis=1)   
            
            # recompute centroids
            clusters = [np.mean(X[cluster_assignment==i], axis=0) for i in range(len(clusters))]
            clusters = np.r_[[np.mean(X, axis=0) if np.alltrue(np.isnan(c)) else c for c in clusters]]      

        self.cluster_centers_ = clusters
            
    def predict(self, X):
        # assign each data point to closest centroid
        d = np.r_[[[self.metric(c, x) for x in X] for c in self.cluster_centers_]].T
        y = np.argmin(d, axis=1)   
        return y
    
    
    def plot_clusters(self,X):
        y = self.predict(X)
        n_clusters = len(np.unique(y))
        cmap = plt.cm.bwr if n_clusters==2 else plt.cm.plasma
        cmap((y*255./(n_clusters-1)).astype(int))
        for i in np.unique(y):
            col = cmap((i*255./(n_clusters-1)).astype(int))
            Xr = X[y==i]
            plt.scatter(Xr[:,0], Xr[:,1], color=col, label="cluster %d"%i, alpha=.5)
        plt.scatter(self.cluster_centers_[:,0], self.cluster_centers_[:,1],marker="x", lw=5, s=200, color="black")
        plt.legend()    
        plt.xlabel("anchura del trilobite")
        plt.ylabel("largo del trilobite");
```

```python
km = KM(n_clusters=2)
km.fit(X)
km.plot_clusters(X)
```

```python
km = KM(n_clusters=5)
km.fit(X)
km.plot_clusters(X)
```

## we now use our own metrics

```python
def my_metric0(x,y):
    return np.abs(x[0]-y[0])

km = KM(n_clusters=5, metric=my_metric0)
km.fit(X)
km.plot_clusters(X)
```

```python
def my_metric1(x,y):
    return np.abs(x[1]-y[1])


km = KM(n_clusters=5, metric=my_metric1)
km.fit(X)
km.plot_clusters(X)
```

```python
def my_metric2(x,y):
    return .5*np.abs(x[0]-y[0]) + np.abs(x[1]-y[1])


km = KM(n_clusters=5, metric=my_metric2)
km.fit(X)
km.plot_clusters(X)
```
